# 5.15 — Inspecting Feature Distributions Before Modeling

Before modeling, inspect how each feature behaves:

- Numerical features: range, percentiles, skew, outliers
- Categorical features: value counts, rare/unexpected levels
- Compare distributions across target classes (when applicable)

This repo includes `src/feature_inspection.py` to generate summaries and optionally save plots (if matplotlib is installed).

## 1) Load raw data and show schema
Data loading should not transform or fit anything.

```python
from src.data_loader import load_csv

df = load_csv('data/raw/source_demo_crops_20260321.csv')
print('shape:', df.shape)
print('dtypes:\n', df.dtypes)
df.head()
```

## 2) Use the configured feature lists
Feature lists are defined centrally in `src/config.py` (5.14).

```python
from src.config import (
    ALL_FEATURES,
    CATEGORICAL_FEATURES,
    EXCLUDED_COLUMNS,
    NUMERICAL_FEATURES,
    TARGET_SOURCE_COLUMN,
)

print('NUMERICAL_FEATURES:', NUMERICAL_FEATURES)
print('CATEGORICAL_FEATURES:', CATEGORICAL_FEATURES)
print('EXCLUDED_COLUMNS:', EXCLUDED_COLUMNS)
print('ALL_FEATURES:', ALL_FEATURES)
print('Target source:', TARGET_SOURCE_COLUMN)
```

## 3) Distribution summaries (numeric + categorical)
These summaries help catch skew, outliers, rare categories, and missing values before training.

```python
from src.feature_inspection import inspect_feature_distributions

numeric_summaries, categorical_summaries = inspect_feature_distributions(
    df,
    numeric_features=NUMERICAL_FEATURES,
    categorical_features=CATEGORICAL_FEATURES,
)

numeric_summaries, categorical_summaries
```

## 4) Compare numeric distributions by target (demo)

For the demo dataset we derive a binary target from `yield_kg`.
Then we compare numeric distributions across the two target classes.

```python
from src.feature_inspection import compare_numeric_by_target

y = (df[TARGET_SOURCE_COLUMN] >= df[TARGET_SOURCE_COLUMN].median()).astype(int)
comparison = compare_numeric_by_target(df, numeric_features=NUMERICAL_FEATURES, target=y)
comparison
```
