## Step 1: Define Input Data
The next code cell defines a small sample dataset for demonstration.

```python
crop_data = [
    {'crop': 'Rice', 'expected_price': 2100},
    {'crop': 'Wheat', 'expected_price': 1950},
    {'crop': 'Maize', 'expected_price': 1800}
]
crop_data
```

## Step 2: Perform a Simple Calculation
The next code cell computes the average expected price from the sample data.

```python
avg_price = sum(item['expected_price'] for item in crop_data) / len(crop_data)
print('Average expected price:', round(avg_price, 2))
```

## Step 3: Conclusion

## Step 4: Execution Behavior (Interrupt vs Restart)
This section demonstrates intentional execution, safe interruption of a long-running cell, and kernel restart behavior.

```python
import time
kernel_state = 'initialized'
print('kernel_state set to:', kernel_state)
```

```python
# Long-running cell used to demonstrate safe interrupt.
for i in range(1, 31):
    print(f'Running second {i}/30')
    time.sleep(1)
print('Long-running cell finished.')
```

### Simulated Safe Interrupt
The next cell intentionally raises `KeyboardInterrupt` to demonstrate interrupt behavior without freezing the notebook session.

```python
for i in range(1, 11):
    print(f'Loop step {i}')
    time.sleep(0.2)
    if i == 4:
        raise KeyboardInterrupt('Simulated user interrupt at step 4')
```

```python
try:
    print('kernel_state is available:', kernel_state)
except NameError:
    print('kernel_state is missing. Kernel was restarted; rerun setup cells to restore state.')
```

## Step 5: When to Interrupt vs Restart
Use **Interrupt** when a single cell is running too long, stuck in a loop, or using wrong input, and you want to keep current variables in memory.
Use **Restart Kernel** when notebook state is inconsistent, imports or variables are corrupted, or you need a clean reproducible run from top to bottom.
